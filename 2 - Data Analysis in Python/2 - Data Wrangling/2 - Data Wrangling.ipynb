```python
import pandas as pd
import numpy as np

other_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(other_path, header=None)
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
```

```python
df=df.replace('?',np.NaN)
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
```

```python
df["city-mpg"] = 235/df["city-mpg"]
```

```python
df.rename(columns={"city-mpg":"city-L/100km"},inplace=True)
```

```python
df
```

```python
avgnl =df['normalized-losses'].mean(axis=0)

df['normalized-losses'].replace(np.nan, avgnl)
```

```python
df[["length","width","height"]]
```

```python
#Simple Feature Scaling (dividing by max)
df["length"] = df["length"]/df["length"].max()
```

```python
df["length"].min()
```

```python
#Min-max method (dividing by range)
df["width"] = (df["width"] - df["width"].min())/(df["width"].max() - df["width"].min())
```

```python
df["width"]
```

```python
#Actual standardisation
df["height"]  = (df["height"] - df["height"].mean()) / df["height"].std()
```

```python
df["height"]
```

```python
df["price"] = df["price"].astype(int)
```

```python
df.describe()
```

```python
bins = np.linspace(min(df["price"]),max(df["price"]),4)
```

```python
group_names = ["low","medium","high"]
```

```python
df["price-binned"] = pd.cut(df["price"],bins,labels=group_names,include_lowest=True)
```
